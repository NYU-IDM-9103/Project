```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import keras
from keras.datasets import cifar10
```

```python
# note: need to downgrade numpy pip install --upgrade numpy==1.26.0
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
print("Training data shape:", X_train.shape)
print("Training labels shape:", y_train.shape)
print("Testing data shape:", X_test.shape)
print("Testing labels shape:", y_test.shape)
```

```python
def plot_10_by_10_images(images):

    # figure size
    fig = plt.figure(figsize=(10,10))

    # plot image grid
    for x in range(10):
        for y in range(10):
            ax = fig.add_subplot(10, 10, 10*y+x+1)
            plt.imshow(images[10*y+x])
            plt.xticks(np.array([]))
            plt.yticks(np.array([]))
    plt.show()

plot_10_by_10_images(X_train[:100])
```

```python
# check distribution of train labels
unique_items, counts = np.unique(y_train.ravel(), return_counts=True)
train_dist = dict(zip(unique_items, counts))
train_dist
```

```python
# check distribution of test labels
unique_items, counts = np.unique(y_test.ravel(), return_counts=True)
test_dist = dict(zip(unique_items, counts))
test_dist
```
